```python
for idx,filestr in enumerate(allfitsfiles):
    
    subplotborders = subplotfigs[int(idx % 9)]
    colorbarborder = subplotborders[:]
    colorbarborder[1] = 0.03
    colorbarborder[3] = 0.02
    
    srcname = filestr.split("__")[0]
        
    if (idx % 9) == 0:
        fig = plt.figure(figsize=(18,2))
        
        
    f = aplpy.FITSFigure(filestr, figure=fig, subplot=subplotborders)
    f.show_colorscale(cmap = "viridis")
    
    # turning this on messes up the whole image lol.
    """
    f.add_colorbar()
    f.colorbar.set_box(colorbarborder)
    f.colorbar.set_location('bottom')
    f.colorbar.set_pad(0.0)
    f.colorbar.set_width(0.02)
    """
    
    f.add_beam()
    """
    f.beam.show()
    f.beam.set_corner("bottom left")
    f.beam.set_color("black")
    f.beam.set_edgecolor("black")
    f.beam.set_facebolor("white")
    """
    if (idx % 9) == 0:
        #f.axis_labels.set_xtext('Right Ascension (J2000)')
        #f.axis_labels.set_ytext('Declination (J2000)')
        
        #f.tick_labels.set_xformat('mm:ss')
        #f.tick_labels.set_yformat('mm:ss')
        
        f.add_label(0.1, 0.9, "{}".format(srcname), relative=True,
                   backgroundcolor="pink", size="x-small")
        
        f.axis_labels.hide()
        f.tick_labels.hide()
                
        f.add_scalebar(1 * u.arcsecond)
        #f.scalebar.show(1 * u.arcsecond)
        f.scalebar.set_label('1"')
        #f.scalebar.set_corner("bottom")
        f.scalebar.set_font(size='x-small', weight='medium', stretch='normal', 
                            family='sans-serif', style='normal', variant='normal')
    else:
        f.axis_labels.hide()
        f.tick_labels.hide()
        
    if (idx % 9) == 8:
        # assumes each source has all 8 lines to plot.
        print("{0}\nsaving images for {1}\n{0}".format("#"*30, srcname))
        fig.canvas.draw()
        fig.savefig("{}.png".format(srcname), dpi=400)
```
