```python
df = astropy.io.ascii.read("../data/distances.txt").to_pandas()
df[df['Main'] == 'HOPS-384']
```

```python
df[df['Main'] == 'HOPS-12']
```

```python
df = df[df['Main'].isin(targets)]
df['RA'] = 15 * (df['RAh'] + df['RAm'] / 60 + df['RAs'] / 3600)
df['DE-'] = df["DE-"].apply(lambda x: -1 if x == "-" else 1)
df['Dec'] = df['DE-'] * (df['DEd'] + df['DEm'] / 60 + df['DEs'] / 3600)
df = df[['Main', 'Source', 'RA', 'Dec', 'Dis', 'LBol', 'TBol', 'Class', 'SigmaYSO']]
# df = df.set_index(['Main', 'Source'])
df = df.sort_values(['Main', 'Source']).reset_index(drop=True)
df.to_csv("../data/source_info.csv",index=False)
df
```
