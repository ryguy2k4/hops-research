```python
from paramiko import SSHClient
from scp import SCPClient
import os
```

This notebook is for automating scp

```python
with open("../data/12CO_cubes.txt") as f:
    files = f.readlines()

co12_files = list(map(lambda x: str(x).removesuffix("\n"), files))
cont_files = list(map(lambda x: os.path.join(os.path.split(os.path.split(str(x))[0])[0], 'cont.fits'), co12_files))
```

```python
hostname = "eeyore.astro.illinois.edu"
username = "ryants4"
local_path = "/Volumes/Alpha/Research/data"

with open("./pwd.txt") as f:
    password = f.readline()
```

```python
def copy_remote_files(files):
    ssh = SSHClient()
    ssh.load_system_host_keys()
    ssh.connect(hostname, username=username, password=password)
    print("Session Connected")

    with SCPClient(ssh.get_transport()) as scp:
        for path in files:
            # for cube files
            target_name = os.path.basename(path.split("__")[0])

            # for cont files
            # target_name = str(path).split('/')[5]

            # set up destination location
            target_path = os.path.join(local_path, target_name)
            if not os.path.exists(target_path):
                os.mkdir(target_path)

            # Copy file from remote to local
            scp.get(path, target_path)
            print(f"SCP Successful: {os.path.basename(target_name)}")

    ssh.close()
    print('Session Closed')
```

```python
copy_remote_files(cont_files)
```
